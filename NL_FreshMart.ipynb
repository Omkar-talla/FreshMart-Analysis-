**Sales Revenue is Influenced by Price, Quantity, and Discount:**
We have the following based on the hypothesis

**1.Positive Correlation between Price and Revenue:**

Higher-priced products tend to generate more total sales revenue.

**2.Positive Correlation between Quantity and Revenue:**

More units sold directly boost overall revenue.

**3.Negligible Correlation between Discount and Revenue:**

Discounts have minimal effect on revenue, suggesting discount strategies might not be as effective.

**4.No Correlation between Price and Quantity:**

Customers do not necessarily buy more or fewer units based on product price alone.

**5.No Correlation between Discount and Quantity:**

Increasing discounts doesn't significantly influence how much customers purchase.

The correlation matrix here shows the relationship between price,quantity,discount,total sales revenue and how they are related to each other.
A positive correlation indicates that as one variable increases,the other tends to increase as well, while a negative correlation suggests that as one increases, the other tends to decrease.
For instance,a strong positive correlation between Quantity and Total Sales Revenue would indicate that higher sales volumes directly drive more revenue.
A negative correlation between Discount and Total Sales Revenue might suggest that excessive discounting is hurting overall revenue, possibly because the increase in quantity sold does not fully compensate for the loss in per-unit earnings.
Similarly, a negative correlation between Price and Quantity may reflect that higher-priced products are being purchased less frequently.

```
numeric_cols = Sales_per_revenue[['Price', 'Quantity', 'Discount', 'Total Sales Revenue']]
correlation_matrix = numeric_cols.corr()
print(correlation_matrix)
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix of Sales Metrics")
plt.show()
```

# Key Findings and Recommendations

**Key Findings and Recommendations:**

**1.Product-Level Insights:**

A small subset of products contributes a significant portion of the total sales revenue.
Some low-priced products are selling in high volumes but contributing less to revenue due to high discount rates.

**2.Customer Behavior:**

Top 10 customers by revenue consistently purchase high-value or bulk items, indicating a segment of premium shoppers.
Customers with incomplete profile information e.g., missing city or middle initial are slightly less likely to generate repeat high-value purchases.

**3.Discounts and Revenue Tradeoff:**

Products with higher discount percentages do not always result in proportionally higher quantities sold.
Excessive discounting leads to revenue leakage, especially in mid-range priced products.

**4.City-Level Sales Trends:**

A few cities dominate total revenue. These cities tend to have better customer engagement and possibly more active employees.
Cities with incomplete or missing data show erratic purchase patterns and lower average order values.

**5.Employee Impact:**

Employees with more sales interactions  contribute significantly to revenue generation, indicating value in training or incentivizing top performers.

**Conclusions from the FreshMart Sales Analysis:**

A small number of products drive the majority of revenue, indicating a skewed product performance where only top sellers significantly contribute to total sales.

High discount rates do not always lead to increased sales revenue. In many cases, they reduce profitability without boosting quantity sold proportionally.

Top customers consistently purchase high-priced items, contributing disproportionately to overall revenue, suggesting a loyal high-value customer segment.

Sales are concentrated in a few key cities, which show better performance, possibly due to better infrastructure, marketing efforts, or store presence.

Incomplete customer or city data correlates with lower revenue generation, highlighting the importance of clean, complete data for accurate targeting and analysis.

Employee involvement where trackable affects revenue, with some employees showing consistently higher sales performance than others.

Some product categories underperform despite price adjustments or discounts, indicating potential mismatches with customer demand or seasonality.

Revenue varies significantly across cities and countries, reflecting geographic differences in purchasing behavior, which may inform location-specific strategies.

**Solutions to Business Objectives:**

1**.Optimize Discount Strategies:**

Re-evaluate discount levels for mid-range and high-selling products.
Implement a tiered discount system based on quantity or customer loyalty.

**2.Focus on Top-Performing Cities and Customers:**

Increase marketing efforts in high-revenue cities.
Provide targeted offers to top 10% of customers based on historical purchase value.

**3.Improve Data Completeness:**

Ensure customer and city information is complete during onboarding or purchase.
Use mandatory fields in POS/data entry to avoid incomplete records.

**4.Empower Employees:**

Identify high-performing employees and replicate their strategies across teams.

Offer incentives for employees tied to sales metrics.

**5.Expand High Revenue Product Lines:**

Introduce variants or combo offers of top-selling products.

Discontinue or rebrand underperforming items based on sales analysis.
